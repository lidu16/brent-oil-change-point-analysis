```python
# ============================================================================
# 10. LOAD AND EXPLORE THE EVENTS DATASET
# ============================================================================
# This is for Task 1 - creating a structured dataset of major events

import pandas as pd

# Load the events CSV
events_df = pd.read_csv('../data/events.csv')

print("="*60)
print("EVENTS DATASET LOADED")
print("="*60)
print(f"Total events: {len(events_df)}")
print("\nEvents:")
print(events_df.to_string(index=False))

# Check for any missing values
print("\nMissing values per column:")
print(events_df.isnull().sum())

# Show event count by category
print("\nEvent count by category:")
print(events_df['Category'].value_counts())
```
